```python
import folium

# Your list of locations and their corresponding coordinates
locations = {
    "Burbank": [34.180840, -118.308968],
    "Van Nuys": [34.189857, -118.451355],
    "Hawthorne": [33.919434, -118.353996],
    # Add the rest of your locations here...
}

# Initialize a map centered around Los Angeles, CA
map = folium.Map(location=[34.0522, -118.2437], zoom_start=9)

# Loop through the locations and add each to the map
for location, coordinates in locations.items():
    folium.Marker(
        location=coordinates,
        popup=location,
        icon=folium.Icon(icon="cloud"),
    ).add_to(map)

# Save the map to an HTML file
map.save("map.html")
```

```python
import folium

# Create map centered on Los Angeles
la_map = folium.Map(location=[34.0522, -118.2437], zoom_start=11)

# List of locations with their latitudes, longitudes, and labels
locations = [
    {"label": "Burbank, CA", "latitude": 34.180840, "longitude": -118.308968},
    {"label": "Van Nuys, CA", "latitude": 34.189857, "longitude": -118.451355},
    {"label": "Hawthorne, CA", "latitude": 33.919434, "longitude": -118.353996},
    {"label": "University of Southern California", "latitude": 34.021984, "longitude": -118.284927},
    {"label": "Santa Monica Pier", "latitude": 34.010090, "longitude": -118.496948},
    {"label": "Old Town Pasadena", "latitude": 34.14695, "longitude": -118.14906},
    {"label": "Sunset Strip", "latitude": 34.090698, "longitude": -118.386002},
    {"label": "Griffith Observatory", "latitude": 34.118536, "longitude": -118.300446},
    {"label": "Baldwin Hills Scenic Overlook", "latitude": 34.017583, "longitude": -118.38403},
    {"label": "Rose Bowl Stadium", "latitude": 34.161327, "longitude": -118.167648},
    {"label": "Dodger Stadium", "latitude": 34.073814, "longitude": -118.240784},
    {"label": "Hollywood Bowl", "latitude": 34.112236, "longitude": -118.339432},
    {"label": "Union Station", "latitude": 34.055592, "longitude": -118.233443},
    {"label": "The Broad Museum", "latitude": 34.05446, "longitude": -118.25059}
]

# Add each location to the map
for loc in locations:
    # Create an HTML string for the popup, specifying the font size
    popup_html = f"<div style='font-size: 10pt'>{loc['label']}</div>"
    popup = folium.Popup(popup_html, show=True)
    
    folium.Marker([loc["latitude"], loc["longitude"]], popup=popup).add_to(la_map)



# Display the map
la_map

```
